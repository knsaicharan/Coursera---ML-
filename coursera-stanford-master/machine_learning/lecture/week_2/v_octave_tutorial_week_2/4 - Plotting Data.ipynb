# 1) Plot

```octave_kernel
t = [0:0.01:0.98];
y1 = sin(2*pi*4*t);
```

```octave_kernel
plot(t,y1)
```

```octave_kernel
y2 = cos(2*pi*4*t);
plot(t,y2);
```

```octave_kernel
plot(t,y1);
hold on; % hold on will keep the image, so we can plot the 2nd image on top of the 1st one.
plot(t,y2,'r');
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('my plot')
print -dpng 'myPlot.png'
```

```octave_kernel
close % close the plot
```

# 3) Subplot

```octave_kernel
subplot(1,2,1); % Divides plot a 1x2 grid, access first element
plot(t,y1);
subplot(1,2,2);
plot(t,y2);
axis([0.5 1 -1 1]) % set x range and y range for the figure on the right
```
